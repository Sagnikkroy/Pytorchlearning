```python
class Stud:
    def __init__(self, name, yearofadmission):
        self.name = name
        self.yearofadmission = yearofadmission

    def yearofpassout(self):
        self.passoutyear=self.yearofadmission + 4
        return self.passoutyear

```

```python
student1 = Stud("Sagnik", 2023)
student1.yearofpassout()
print(f"{student1.name} will pass out in the year {student1.passoutyear}")
```
